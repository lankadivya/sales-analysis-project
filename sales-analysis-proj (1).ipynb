```python
import pandas as pd
import os
```

**task #1 - Merge the 12 months of sales data into single Csv file**

reading single csv file of month april

```python
df = pd.read_csv("/kaggle/input/sales-data-csv/Sales_April_2019.csv")
df
```

```python
df.head()
```

```python
files=[file for file in os.listdir('/kaggle/input/sales-data-csv')]
all_months_data=pd.DataFrame()
for file in files:
    df=pd.read_csv('/kaggle/input/sales-data-csv'+'/'+file)
    all_months_data=pd.concat([all_months_data,df])
all_months_data.to_csv("all_data.csv",index=False)
```

**read in updated dataframe**

```python
all_data=pd.read_csv("all_data.csv")
all_data
```

# clean up the data

```python
nan_df=all_data[all_data.isna().any(axis=1)]
nan_df
```

```python
all_data.dropna(inplace=True)
```

**find Or and delete it**

```python
all_data=all_data[all_data["Order Date"].str[0:2]!="Or"]
all_data
```

**convert columns to correct typeQuantity Ordered**

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
```

# Argument data with additional columns

**task #2  :Add month column**

```python
all_data['month']=all_data['Order Date'].str[0:2]
all_data['month']=all_data['month'].astype('int32')
all_data.head()
```

****task #3- add a sales****

by correcting the datatype

```python
all_data['sales']=all_data['Quantity Ordered']*all_data['Price Each']
```

```python
all_data
```

```python
all_data.info()
```

**task 4: Add a city column**

```python
#let's use apply() method
all_data['city']=all_data['Purchase Address'].apply(lambda x:x.split(',')[1])
```

```python
all_data.head()
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(",")[2].split(' ')[1]

all_data['city']=all_data['Purchase Address'].apply(lambda x:get_city(x)+" ("+get_state(x)+")")

all_data.head()
```

# Question 1:what was the best month for sales? How much was earned that month

```python
g=all_data.groupby('month').sum()
g
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.xlabel('months')
plt.ylabel('sales')
plt.bar(months,g['sales'])
plt.xticks(months)
plt.show()
```

# Question 2: which product has highest no of sales?

```python
all_data['Product'].unique()
```

```python
p=all_data.groupby('Product').sum()
p.head()
```

```python
product=[pt for pt,df in all_data.groupby('Product')]
plt.xlabel('product')
plt.ylabel('sales')
plt.bar(product,p['sales'])
plt.xticks(product,rotation="vertical",size=6)
plt.show()
```

# Question 3: which city has the highest no of sales?

```python
c=all_data.groupby('city').sum()
c
```

```python
cities=[city for city,df in all_data.groupby('city')]
plt.xlabel('city')
plt.ylabel('sales')
plt.bar(cities,c['sales'])
plt.xticks(cities,rotation="vertical",size=6)
plt.show()
```

# Question 4: what time should we display the advertisements to maximize the livelihood of customer's buying product?

**convert the order date to python datatime format**

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

```python
all_data['hour']=all_data['Order Date'].dt.hour
all_data['minute']=all_data['Order Date'].dt.minute
```

```python
all_data.head()
```

```python
hours=[hour for hour,df in all_data.groupby('hour')]
plt.plot(hours,all_data.groupby(['hour']).count())
plt.xticks(hours)
plt.grid()
plt.title('Timing Analysis')
plt.xlabel("hour")
plt.ylabel('no of orders')
plt.show()
# my recommendation is around 11 am (11) and 7 pm (19) is the best time to advertise
```
